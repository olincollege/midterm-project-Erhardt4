```python

r = csv.reader(open("EIU_Democracy_Index.csv", "r"))
countryList = []
#Countries List
counter = True
for row in r:
    if counter:
        yearList = row[1:]
        counter = False
    countryList.append(row[0])
countryList= countryList[1:]
#print(countryList)
#Years List


for i in range(len(yearList)):
    yearList[i] = int(yearList[i])
print(yearList)
```

```python
import statistics
countriesAboveThreshold = []
threshold = .08#0.0632
for i in range(164):
    temp = EIU_dataframe.iloc[i,1:14]
    if statistics.pstdev(temp) > threshold:
        countriesAboveThreshold.append(i)
print(countriesAboveThreshold)
#findPopulationStandardDev(temp)

chosenCountriesNames = []
for index in countriesAboveThreshold:
    temp = EIU_dataframe.iloc[index,1:14]
    chosenCountriesNames.append(countryList[index])
    plt.plot(yearList, temp, marker =  '+')

plt.xlabel("Years")
plt.ylabel("EIU Democracy Index Value")
plt.xlim(2006, 2024)
plt.legend(chosenCountriesNames, loc = 'best')

#print(temp)
```
